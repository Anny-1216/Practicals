```python
def clean_data(data):
  # Filter users with empty names
  data["users"] = [i for i in data["users"] if i["name"] != ""]
  # Filter users with no friends
  data["users"] = [i for i in data["users"] if len(i["friends"]) > 0]

  # Make friends list unique for each user
  for user in data["users"]:
    user["friends"] = list(set(user["friends"]))

  # Filter pages to keep only unique IDs (keeping the first occurrence)
  seen_page_ids = set()
  unique_pages = []
  for page in data["pages"]:
      if page["id"] not in seen_page_ids:
          unique_pages.append(page)
          seen_page_ids.add(page["id"])
  data["pages"] = unique_pages

  return data

data = json.load(open("new.json"))
cleaned_data = clean_data(data)
json.dump(cleaned_data, open("new.json", "w"),indent=4)
print("Data Cleaning Sucessful")
```
